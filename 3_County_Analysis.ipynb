#  How do child fatalities due to abuse and neglect vary across different counties in Texas?

## Methodology  

To analyze how child fatalities due to abuse and neglect vary across different regions in Texas, the following steps were taken:  

1.**Aggregate fatalities by region** – Summing the total number of abuse- and neglect-related child fatalities for each region.  
2.**Sort regions by fatalities** – Identifying which regions have the highest and lowest child fatalities.  
3.**Visualize the regional differences** – Creating a **bar chart** to compare fatalities across different Texas regions.  
4.**Identify key patterns** – Observing whether certain regions consistently report higher fatalities and discussing possible contributing factors.  


## Import Libraries and Data

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('/Users/kubra/Desktop/Pyhton/python projects /abuse_neglected_child_fatalities /cleaned_data.csv')

# Display the first few rows
df.head()
     
```

```python
# Check unique regions in the dataset
df['Region'].unique()
```

## Aggregate Child Fatalities by Region  

To understand the geographic distribution of child fatalities, the data needs to be grouped by region. This helps in identifying which areas have the highest number of fatalities and whether certain regions are consistently more affected than others.


```python
# Group by Region and sum fatalities
fatalities_by_region = df.groupby('Region')['Abuse Neglect Fatalities'].sum().reset_index()

# Rename columns for clarity
fatalities_by_region.columns = ['Region', 'Total Fatalities']

# Display the aggregated data
fatalities_by_region
```

## Sort Regions by Fatalities  

After aggregating the number of child fatalities by region, the next step is to sort the data. Sorting allows for a clearer understanding of which regions have the highest and lowest fatalities, making it easier to identify high-risk areas and compare regional differences effectively.


```python
# Sort regions from highest to lowest fatalities
fatalities_by_region = fatalities_by_region.sort_values(by='Total Fatalities', ascending=False)

# Display top regions
fatalities_by_region
```

## Visualizing Child Fatalities by Region  

After sorting the data, the next step is to visualize the distribution of child fatalities across different regions in Texas. A bar chart is used to compare the total fatalities in each region, highlighting areas with the highest and lowest numbers of reported cases.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(12, 6))

# Create a bar chart
sns.barplot(
    data=fatalities_by_region,
    x="Total Fatalities",
    y="Region",
    palette="Blues_r"
)

# Add labels and title
plt.xlabel("Total Fatalities")
plt.ylabel("Region")
plt.title("Child Fatalities Due to Abuse and Neglect by Region (Texas, FY2015-FY2024)")
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Save the figure
plt.savefig("child_fatalities_by_region.png", dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
